```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
```

```python
import matplotlib.pyplot as plt
```

```python
data = mnist.load_data()
(x_train, y_train), (x_test, y_test) = data
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
fig,axis=plt.subplots(2,5)
for i in range(10):
    axis[i//5,i%5].imshow(x_train[i],cmap='gray')
```

# adding noise

```python
import numpy as np
```

```python
noise=np.random.uniform(0,1,size=x_train.shape)
noise2=np.random.uniform(0,1,size=x_test.shape)
```

```python
fig,axis=plt.subplots(2,5)
for i in range(10):
    axis[i//5,i%5].imshow(noise[i],cmap='gray')
```

## adding noise to x_train and x_test

```python
noise_x_train=x_train+noise
noise_x_test=x_test+noise2
```

```python
fig,axis=plt.subplots(2,5)
for i in range(10):
    axis[i//5,i%5].imshow(noise_x_train[i],cmap='gray')
```

## modelimg

```python
model=Sequential()
model.add(Flatten(input_shape=x_train[1].shape))
model.add(Dense(132,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(132,activation='relu'))
model.add(Dense(28*28,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
x_train2=x_train.reshape(-1,28*28)
```

```python
model.fit(noise_x_train,x_train2,validation_split=0.2,epochs=10)
```

```python
pred=model.predict(noise_x_test)
```

```python
pred2_1=pred.reshape(-1,28,28)
```
