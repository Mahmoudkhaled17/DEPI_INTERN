```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/emails.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['spam'])
```

```python
plt.pie(df['spam'].value_counts(),labels=['not spam','spam'],autopct='%1.1f%%')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])
```

```python
print(X[0,:])
```

```python

import pandas as pd
df_x = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
df_x
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
from sklearn.model_selection import train_test_split
(x_train,x_test,y_train,y_test)=train_test_split(df_x,df['spam'],test_size=0.2,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
  from sklearn.metrics import confusion_matrix
  from sklearn.metrics import classification_report
  y_pred=model.predict(x_test)
  print(confusion_matrix(y_test,y_pred))
  import seaborn as sns
  sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
  print(classification_report(y_test,y_pred))
```
