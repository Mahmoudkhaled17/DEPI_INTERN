```python
class studient:
  mydict={}
  def __init__(self,name=None,age=None,grade=None,id=None,email=None):
    self.name=name
    self.age=age
    self.grade=grade
    self.id=id
    self.email=email

    if name is None or age is None or grade is None or id is None or email is None:
     raise ValueError("All fields are required")

    if not isinstance(name,str):
      raise TypeError("Name must be a string")
    if not isinstance(email,str):
      raise TypeError("Email must be a string")
    if not isinstance(age,int):
      raise TypeError("Age must be an integer")
    if not isinstance(grade,int):
      raise TypeError("Grade must be an integer")
    if not isinstance(id,int):
      raise TypeError("ID must be an integer")

    studient.mydict[self.name]=self.grade

  def __str__(self):
    return f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}, ID: {self.id}, Email: {self.email}"

  @classmethod
  def get_max_grade(cls):
    maximum=list(cls.mydict.values())[0]
    for i in cls.mydict.values():
      if i>maximum:
        maximum=i
    return maximum


  @classmethod
  def max_grade(cls):
    return max(cls.mydict.values())

  @classmethod
  def average(cls):
    return sum(cls.mydict.values())/len(cls.mydict)



```
