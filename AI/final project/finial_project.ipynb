```python
# Install Gradio
!pip uninstall gradio -y
!pip install gradio

import gradio as gr
import cv2
from PIL import Image
import numpy as np
import torch
# Load your YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/exp/weights/best.pt', force_reload=True)

def predict_image(image):
  """
  Predicts objects in an image using the YOLOv5 model.
  """
  results = model(image)
  # Process results (e.g., draw bounding boxes, get labels)
  annotated_frame = np.squeeze(results.render())
  return Image.fromarray(annotated_frame)

# Create a Gradio interface
# Use gr.Image() instead of gr.inputs.Image() and gr.outputs.Image()
iface = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="Space Debris Detector",
    description="Upload an image to detect space debris."
)

# Launch the interface
iface.launch()
```

```python

from roboflow import Roboflow
import gradio as gr
import cv2
from PIL import Image
import numpy as np
import torch

def predict_video(video_path):
    """
    Predicts objects in a video using the YOLOv5 model.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise IOError("Cannot open webcam")

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    out = cv2.VideoWriter('output.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)
        annotated_frame = np.squeeze(results.render())
        out.write(annotated_frame)

        # Display the annotated frame (optional)
        # cv2.imshow('YOLOv5 Detection', annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()
    return 'output.avi'


iface = gr.Interface(
    fn=predict_video,
    inputs=gr.Video(sources="upload"), # changed from source to sources
    outputs=gr.Video(),
    title="Space Debris Detector",
    description="Upload a video to detect space debris."
)

iface.launch()
```
