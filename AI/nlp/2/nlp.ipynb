```python
from datasets import load_dataset
```

```python
df=load_dataset('asas-ai/Arabic_Sentiment_Twitter_Corpus')
```

```python
df
```

```python
tweet = list(df['train']["tweet"]) + list(df['test']["tweet"])
label = list(df['train']["label"]) + list(df['test']["label"])
```

```python
tweet[0:10]
```

```python
import re
```

```python
text_train_2 = []
arabic_pattern = re.compile(r'[\u0600-\u06FF]+')

for text in tweet:
  cleaned = " ".join(arabic_pattern.findall(text))
  text_train_2.append(cleaned)


print(text_train_2[0:10])

```

```python
!pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
arb_stopwords = set(stopwords.words("arabic"))
print(arb_stopwords)
```

```python

text_no_stopwords = []

for text in text_train_2:
  words = text.split()
  filtered_words = [word for word in words if word not in arb_stopwords]
  text_no_stopwords.append(" ".join(filtered_words))

```

```python
text_no_stopwords[0:10]
```

```python
!pip install qalsadi
from qalsadi import lemmatizer
lemmer = lemmatizer.Lemmatizer()

clean_text_lemmatized = []
clean_text_lemmatized_test = []
for text in text_no_stopwords:
  words = text.split()
  lemmatized_words = [lemmer.lemmatize(word) for word in words]
  clean_text_lemmatized.append(" ".join(lemmatized_words))
```

```python
clean_text_lemmatized[0:10]
```

```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(clean_text_lemmatized)
import pandas as pd
data_v = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
data_v
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv_tfidf = TfidfVectorizer()
X_tfidf = cv_tfidf.fit_transform(clean_text_lemmatized).toarray()
data_tf=pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out())

```
